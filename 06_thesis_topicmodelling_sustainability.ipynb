```python
import textwrap
import math
for i in range(0,20):
    publication_number_EPO = patents_text.publication_number_EPO.iloc[i]
    title = patents_text.title.iloc[i]
    abstract = patents_text.abstract.iloc[i]
    description = patents_text.description.iloc[i]
    claims = patents_text.claims.iloc[i]

    # Check if the description is NaN or not a string
    if isinstance(description, float) and math.isnan(description):
        description = "Description is not available."
    elif not isinstance(description, str):
        description = str(description)

    # Print title, abstract, and description with wrapping
    print(textwrap.fill(f"Publication Number: {publication_number_EPO}", width=80))
    print(textwrap.fill(f"Title: {title}", width=80))
    #print(textwrap.fill(f"Abstract: {abstract}", width=80))
    print(textwrap.fill(f"Description: {description[:3000]}", width=80))
    print("\n")
```
